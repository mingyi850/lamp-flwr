```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'Results.xlsx'
data = pd.read_excel(file_path, sheet_name=None)

# Extracting the data for each ROUGE metric
rouge_1_data = data['Rouge-1']
rouge_2_data = data['Rouge-2']
rouge_L_data = data['Rouge-L']
```

```python
for sheet in data:
    data[sheet].columns = ['Dataset'] + data[sheet].columns[1:].tolist()
    data[sheet] = data[sheet].set_index('Dataset')

# Plotting each ROUGE metric again
for rouge_data, title in zip([data['Rouge-1'], data['Rouge-2'], data['Rouge-L']],
                             ['ROUGE-1 Scores', 'ROUGE-2 Scores', 'ROUGE-L Scores']):
    plt.figure(figsize=(5, 5))
    for dataset in rouge_data.index:
        plt.plot(rouge_data.columns, rouge_data.loc[dataset], label=dataset)
    plt.title(title)
    plt.xlabel('Staleness')
    plt.ylabel('Score')
    plt.legend()
    plt.grid(True)
    plt.savefig(title+'.png')
```

```python
file_path = 'staleness_0.out'

# Read the file content
with open(file_path, 'r') as file:
    content = file.read()

# Extract texts between 'prediction: [' - ']'
import re
extracted_texts = re.findall(r"prediction: \['(.*?)'\]", content.replace('\n', ''))

extracted_texts = [i.replace('[CLS] ', '') for i in extracted_texts]
extracted_texts = [i.replace('[SEP]', '') for i in extracted_texts]
extracted_texts = [i.replace("'", '') for i in extracted_texts]
extracted_texts = [i for i in extracted_texts if len(i)<200]
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.animation import PillowWriter
from PIL import Image, ImageDraw, ImageFont
import numpy as np
import textwrap

# Text data
texts = extracted_texts

def text_to_wrapped_image(text, image_size=(1000, 50), font_size=15, line_width=160):
    # Create an image with white background
    img = Image.new('RGB', image_size, color='white')
    d = ImageDraw.Draw(img)
    font = ImageFont.load_default()

    # Wrap the text
    wrapped_text = textwrap.fill(text, width=line_width)

    # Add text to image
    d.multiline_text((10, 10), wrapped_text, fill=(0, 0, 0), font=font)
    return img

# Generate images for each text and save as a sequence
images = [text_to_wrapped_image(text) for text in texts]

# Save the sequence as a GIF
gif_path = 'Staleness_0.gif'
images[0].save(gif_path, save_all=True, append_images=images[1:], optimize=False, duration=500, loop=1)
```
